## Correlação Linear

### Exemplo 1:
Deseja-se testar o efeito do peso do veículo sobre sua distancia de frenagem em uma superficie molhada.

### Experimento:
Foram selecionados 8 veículos e observados seus pesos (em libras) e suas respectivas distancias de frenagem (em pes)
| X (libras) | 5720 | 4050 | 6130 | 5000 | 5010 | 4270 | 5500 | 5550 |
|------------|------|------|------|------|------|------|------|------|
| Y (pes) | 3.78 | 2.43 | 4.63 | 2.88 | 3.25 | 2.76 | 3.42 | 3.51 |

#### Questoes de interesse:
 - Existem diferencas significativas entre os valores da distancia de frenagem devido ao peso do veiculo?
 - De que forma podemos expressar $DistanciaFrenagem = f(PesoVeiculo)$?
    - E como podemos fazer inferencias sobre os parametros envolvidos na funçao
    

### Diagrama de dispersão
plotagem dos pontos (X,Y)

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#Criando os arrays de dados
X = np.array([5720, 4050, 6130, 5000, 5010, 4270, 5500, 5550])
Y = np.array([3.78, 2.43, 4.63, 2.88, 3.25, 2.76, 3.42, 3.51])

#Colocando os arrays em um DataFrame
df1 = pd.DataFrame({'X' : X, 'Y': Y})
```

#### Plotando Diagrama de dispersão

```python
# Plotando um grafico de dispersão dos valores informados
plt.scatter(df1['X'], df1['Y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Dispersão')
```

#### Modelo de regressão linear

```python
# Ajustar o modelo de regressão linear
model1 = LinearRegression()
model1.fit(df1[['X']], df1['Y'])

# Plotando um grafico de dispersão dos valores informados
plt.scatter(df1['X'], df1['Y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Dispersão')

# Plotar a linha de regressão
plt.plot(df1['X'], model1.predict(df1[['X']]), color='red')
```
